```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns

os.chdir('./data')
init_data = pd.read_csv('kc_house_data.csv')
os.chdir('./')

init_data.head()
```

```python
for column in init_data.columns:
    nan_total = init_data[column].isna().sum()
    if nan_total > 0:
        print(f'''{column} has {nan_total} nan vlaues in it, accounting for {round(100*nan_total/init_data.shape[0], 2)}% of that feature.''')
    else:
        continue
```

```python
potential_cat = []
for column in init_data.columns:
    unique = init_data[column].nunique()
    if unique <= 20:
        print(f'{column} only has {unique} values. May be categorical.')
        potential_cat.append(column)
    else:
        continue
```

```python
pot_cat_dict = {}
for n in potential_cat:
    pot_cat_dict[n] = list(init_data[n].unique())

pot_cat_dict
```

```python
# At this point, I decide to drop a few columns. Some are being dropped because I am behind and
# the project notes recommended that I drop some specific columns if I were behind (those are
# view, date, sqft_basement, and yr_renovated, though the last also had a larger percentage
# of NaN values). The other column I'm dropping at this point is the waterfront column. This
# will remove one of the categorical features, as well as deal with the NaN values in the column.

init_data2 = init_data.drop(['view', 'date', 'yr_renovated', 'waterfront', 'sqft_basement']
                            , axis = 1)
init_data2.columns
```

```python
categoricals = []
continuous = []

for column in init_data2:
    if column in potential_cat:
        categoricals.append(column)
    else:
        continuous.append(column)
    
categoricals, continuous
```

```python
# Time to double check the assignment of categorical status for all features.

fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(22,5))

for xcol, ax in zip(categoricals, axes):
    init_data2.plot(kind='scatter', x=xcol, y='price', ax=ax, alpha=0.7, color='pink')
```

```python
# Identified a possible outlier in the bedrooms feature, will revist it later

fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(18,18))

for xcol, ax in zip(continuous, axes.flatten()):
    init_data2.plot(kind='scatter', x=xcol, y='price', ax=ax, alpha=0.7, color='pink')
```

```python
# I can see that bathrooms and zipcode are both categorical.I've also identified possible
# outliers in sqft_lot15, and the house ID feature can most likely be dropped.

categoricals.append('zipcode')
categoricals.append('bathrooms')
continuous.remove('zipcode')
continuous.remove('bathrooms')
continuous.remove('id')

init_data2 = init_data2.drop('id', axis = 1)

categoricals, continuous
```

```python
# Lets deal with the categorical values. Time to make some dummies!

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(16,10), sharey=True)

for col, ax in zip(categoricals, axes.flatten()):
    (init_data2.groupby(col).mean()['price'].sort_values().plot.bar(ax=ax))
    ax.set_title(col)
    
fig.tight_layout()
```

```python
# Zipcode seems excessive, so I'm going to remove that now, as well

init_data2 = init_data2.drop('zipcode', axis=1)
categoricals.remove('zipcode')
```

```python
import pygeohash as gh

# King County, Washington (verified with lat/long) has a land area of 5480 square km. In
# order to eventually create enough different geocashe dummy values, but not too many, I'll
# check on a few different lengths of geohash's to see which will provide a good amount of
# eventual dummy values


for n in range(1, 8):
    needs_pre[f'geohash_{n}'] = needs_pre.apply(lambda x: gh.encode(x.lat, x.long,
                                                                   precision=n), axis=1)
    unique = needs_pre[f'geohash_{n}'].nunique()
    print(f'A geohash of precision={n} provides {unique} unique values')
    
needs_pre.head()
```

```python
# Time to dummify the geohash values. Clearly I should have converted the lat/long values
# to geohash values much earlier in this process, but here we are. I'm going to use the geohash
# values with precision=6 and remove the rest, as well as lat/long from the needs_pre df.

needs_pre = needs_pre.drop(['geohash_1',
                            'geohash_2',
                            'geohash_3',
                            'geohash_4',
                            'geohash_5',
                            'geohash_7',
                            'lat', 'long'], axis=1)

geo_dummies = pd.get_dummies(needs_pre['geohash_6'], prefix='geo', drop_first=True)
dummies = pd.concat([dummies, geo_dummies], axis=1)

display(needs_pre.head())
dummies.head()
```

```python
needs_pre.drop('geohash_6', axis=1, inplace=True)
```

```python
# Next to check for multicollinearity

multicorr = needs_pre.corr().abs().stack().reset_index().sort_values(0, ascending=False)
multicorr['pairs'] = list(zip(multicorr.level_0, multicorr.level_1))
multicorr.set_index(['pairs'], inplace=True)
multicorr.drop(['level_1', 'level_0'], axis=1, inplace=True)
multicorr.columns= ['cc']
multicorr.drop_duplicates(inplace=True)
multicorr[(multicorr.cc>.75) & (multicorr.cc<1)]
```

```python
# With the exception of yr_built, there is a plausible linear relationship in the data for the
# three other variables. For sqft_lot and sqft_lot 15 I need to check to see how many values
# Each has close to zero, and how many possible outliers there are. Once identified, if I want
# to remove those values from the data pool, I'll need to merge the needs_pre df with the
# dummies df before removing them

for column in npre:
    print(column.upper())
    print('----------------------------------------------')
    print(f'{column} minimum is {needs_pre[column].min()}')
    print(needs_pre[column].nsmallest(10).value_counts())
    print(f'{column} maximum is {needs_pre[column].max()}')
    print(needs_pre[column].nlargest(10).value_counts())
    print('----------------------------------------------')
```
